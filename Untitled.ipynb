Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
columns_to_scale = ['Time','Amount']

df_scaled = df.copy()
df_scaled[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
```

```python
from sklearn.ensemble import IsolationForest

# Fit Isolation Forest model
iso_forest = IsolationForest(contamination=0.001, random_state=42)
df['anomaly'] = iso_forest.fit_predict(df.drop('Class', axis=1))
```

```python
df['anomaly'].value_counts()
```

# Split Data into Features and Target

```python
X = df.drop(['Class', 'anomaly'], axis=1)
y = df['Class']
```

Train-test Split and Handle Imbalance with SMOTE

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

# Model Training - 

Train Logistic Regression model

```python
import warnings

# Ignore all warnings
warnings.filterwarnings('ignore')
```

```python
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train_smote, y_train_smote)
```

```python

y_pred_log_reg = log_reg.predict(X_test)
print("Logistic Regression Accuracy: ", accuracy_score(y_test, y_pred_log_reg))
print(classification_report(y_test, y_pred_log_reg))
```

# Train Decision Tree model

```python
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train_smote, y_train_smote)
```

```python
y_pred_tree = tree_model.predict(X_test)
print("Decision Tree Accuracy: ", accuracy_score(y_test, y_pred_tree))
print(classification_report(y_test, y_pred_tree))
```

# Neural Networks (MLPClassifier)

```python
# Train Neural Network model
mlp_model = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
mlp_model.fit(X_train_smote, y_train_smote)
```

```python
# Make predictions and evaluate
y_pred_mlp = mlp_model.predict(X_test)
print("Neural Network Accuracy: ", accuracy_score(y_test, y_pred_mlp))
print(classification_report(y_test, y_pred_mlp))
```
